```python
from skimage import io
import numpy as np
import os

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def tileImage(img, divisor):
    M = img.shape[0]//divisor
    N = img.shape[0]//divisor
    tiles = np.asarray([img[x:x+M,y:y+N] for x in range(0,img.shape[0],M) for y in range(0,img.shape[1],N)])
    return(tiles)
```

```python
tiledPhase1 = tileImage(phaseImg1, 8)
tiledMasks1 = tileImage(maskImg1, 8)
tiledPhase2 = tileImage(phaseImg2, 8)
tiledMasks2 = tileImage(maskImg2, 8)
```
